#### Categrical Data

```python
plt.style.use('ggplot')
plt.style.use('dark_background')
df['Brand of Vehicle'].value_counts().plot(kind='barh',
                               color = 'pink',
                               grid = True,
                               width = 0.6,
                               title = 'Bar Plot for Brand of Vehicle',
                               figsize=(8,5),
                               ylabel='Brand of Vehicle',
                               xlabel = 'No of Vehicles')
```

* From the above bar plot we can see there is more vehicles from Maruti Brand more than 175 count
* And there is minimum cars of KIA, JEEP, SKODA Brand

```python
plt.style.use('ggplot')
plt.style.use('dark_background')
df['Name of City'].value_counts().plot(kind='barh',
                               color = 'blue',
                               grid = True,
                               width = 0.6,
                               title = 'Bar Plot for City of Cars',
                               figsize=(8,5),
                               ylabel='Name of City',
                               xlabel = 'No of Vehicles')
```

* From the above bar plot we can see there is more vehicles in citys like PATNA, KOLKATA, JAIPUR, AGRA, CHENNAI, BENGALURU, COIMBATORE, SURAT, KOCHI

```python
df['Fuel type'].value_counts().plot(kind='bar',
                               color = ['pink','white'],
                               grid = True,
                               width = 0.3,
                               title = 'Bar Plot for Fuel type',
                               figsize=(6,4),
                               xlabel='Fuel type',
                               ylabel = 'No of Vehicles')
```

* From the above plot we can see there is most of the vahicle having Fuel type of Petrol near about 330 count
* And less vehicles having fuel type of Diesel near about 80 count

```python
df['Transmission'].value_counts().plot(kind='bar',
                               color = ['yellow','green'],
                               grid = True,
                               width = 0.3,
                               title = 'Bar Plot for Transmission type',
                               figsize=(6,4),
                               xlabel='Transmission type',
                               ylabel = 'No of Vehicles')
```

* From the above plot we can see there is most of the vahicle having Transmission type of Manual near about 360 count
* And less vehicles having Transmission of Automatic near about 50 count

```python
plt.style.use('classic')
plt.style.use('ggplot')
(df['Transmission'].value_counts()/len(df['Transmission'])*100).plot(kind='pie',
                               figsize = (6,5),
                               legend = True,
                               autopct = '%1.1f'+' Percent' ,                   
                               ylabel = 'Transmission',
                               grid = True,
                               title = 'Pie chart for Transmission type in percentage')
```

* From above Pie chart we can see the percentages of vehicles transmission
* There is 86.8 percent of vehicles having Manual Transmission
* There is 13.2 percent of vehicles having Automatic Transmission

```python
plt.style.use('classic')
plt.style.use('ggplot')
(df['Fuel type'].value_counts()/len(df['Fuel type'])*100).plot(kind='pie',
                               figsize = (6,5),
                               legend = True,
                               autopct = '%1.1f'+' Percent' ,                   
                               ylabel = 'Fuel type',
                               grid = True,
                               title = 'Pie chart for Fuel type in percentage')
```

* From above Pie chart we can see the percentages of vehicles Fuel type
* There is 80.9 percent of vehicles having Petrol Fuel type
* There is 19.1 percent of vehicles having Diesel Fuel type

#### BiVariate Analysis
#### Categrical & Categgrical Data

```python
plt.style.use('ggplot')
plt.style.use('dark_background')
ax = sns.countplot(data = df , y = 'Brand of Vehicle' , hue = 'Transmission',saturation = 0.75)
for i in ax.containers:
    ax.bar_label(i)
```

* From above plot we can see Highest vehicle coming from Maruti which is having Manual transmission
* second highest vehicle coming from Hyundai which is having Manual transmission

```python
plt.style.use('classic')
plt.style.use('ggplot')
plt.style.use('dark_background')
plt.figure(figsize=(8,6))
ax = sns.countplot(data = df , y = 'Brand of Vehicle' , hue = 'Fuel type')
for i in ax.containers:
    ax.bar_label(i)
```

* From above plot we can see Highest vehicle coming from Maruti which is running on petrol near about 153 counts
* second highest vehicle coming from Hyundai which is running on petrol near about 106 counts
